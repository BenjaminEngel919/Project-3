```python
basics = pd.read_csv(basics_url, sep='\t', low_memory=False)

ratings = pd.read_csv(ratings_url, sep='\t', low_memory=False)

akas = pd.read_csv(akas_url, sep='\t', low_memory=False)
```

**Clean the Data - Basics**

Replace \N with Nan

```python
basics = basics.replace({'\\N':np.nan})
```

```python
basics[basics.eq('\\N').any(1)]
```

```python
basics.head()
```

Remove movies with Null runtime.

```python
basics.dropna(subset=['runtimeMinutes'], inplace=True)
```

```python
basics['runtimeMinutes'].isna().sum()
```

Remove movies with Null for genre.

```python
basics.dropna(subset=['genres'], inplace=True)
```

```python
basics['genres'].isna().sum()
```

Keeping on full length movies.

```python
basics['titleType'].unique()
```

```python
movie = basics['titleType'] == 'movie'
basics = basics[movie]
basics['titleType'].unique()
```

```python
basics.head()
```

Keeping movies from 2000 to 2021.

```python
basics['startYear'].info()
```

```python
basics['startYear'] = basics['startYear'].astype('float')
```

```python
years = (basics['startYear'] >= 2000) & (basics['startYear'] <=2021)
basics = basics[years]
basics.head()
```

Eliminate movies with Documentary in genre.

```python
# Exclude movies that are included in the documentary category.
is_documentary = basics['genres'].str.contains('documentary',case=False)
basics = basics[~is_documentary]
```

```python
basics
```

**Clean the Data - AKAS**

```python
akas.head()
```

Replacing \N with NaN.

```python
akas.replace({'\\N':np.nan}, inplace=True)
```

```python
akas.head()
```

USA movies only.

```python
usamovies= akas['region'] == 'US'
```

```python
akas = akas[usamovies]
akas
```

```python
# Filter the basics table down to only include the US by using the filter akas dataframe
keepers =basics['tconst'].isin(akas['titleId'])
keepers
```

```python
basics = basics[keepers]
basics
```
